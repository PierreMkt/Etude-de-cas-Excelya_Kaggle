## Loading data & Preliminary analysis

### Handling irrelevant variables and missing values

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
datasets = [df_train,df_test]
df0 = df_train
```

```python
for df in datasets :
    print(df.count(),'\n') #count the number on non NaN rows

#There are missing values in the Age, Cabin, Fare and Embarked columns
```

'Cabin number' and 'Port of Embarkation' might be relevant regarding the socio economical class of the passengers, but the Cabin variable has too many missing values so we choose to delete it.

```python
for df in datasets :
    display(df.describe(include='all'))
```

Not all variables are relevant to predict survival so we delete the variables that are obviously useless, such as 'PassengerId', 'Name', 'Cabin' and 'Ticket number'.
We use the age median and the port of embarkation mode to replace the missing values.

```python
for df in datasets :
    dropCol = ['PassengerId','Name','Ticket', 'Cabin']
    df.drop(dropCol, axis=1, inplace = True)
    
    df['Age'].fillna(df['Age'].median(), inplace = True)
    df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
    df['Fare'].fillna(df['Fare'].median(), inplace = True)

    print(df.count(),'\n')
    display(df.head())
```

### OneHot Encoding

We OneHot encode the 'Sex' and 'Embarked' variables.
The categorical variable Pclass is already encoded

```python
dfDummiesSex = pd.get_dummies(df_train['Sex'], prefix = 'Sex')
dfDummiesPort = pd.get_dummies(df_train['Embarked'], prefix = 'Port')
dfDummies = pd.concat([dfDummiesSex, dfDummiesPort], axis=1)

df_train = pd.concat([df_train,dfDummies],axis=1)
df_train.drop(['Sex','Embarked'], axis=1, inplace = True)
display(df_train.head())
```

```python
dfDummiesSex = pd.get_dummies(df_test['Sex'], prefix = 'Sex')
dfDummiesPort = pd.get_dummies(df_test['Embarked'], prefix = 'Port')
dfDummies = pd.concat([dfDummiesSex, dfDummiesPort], axis=1)

df_test = pd.concat([df_test,dfDummies],axis=1)
df_test.drop(['Sex','Embarked'], axis=1, inplace = True)
display(df_test.head())
```
